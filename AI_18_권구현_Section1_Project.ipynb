````python
YearQ = df.query('Year <= 1900')
print(YearQ.Year.sort_values())
YearQ20 = YearQ.query('Year <= 23')
print(YearQ20.Year.sort_values())
YearQ19 = YearQ.query('Year > 23')
print(YearQ19.Year.sort_values())
#확인 결과, 23보다 작은 이상치는 앞에 20을 더해주고 23보다 큰 이상치는 19를 더해주면 해결될것이라고 판단
````

````python
def fix_Year(Year):
    if Year < 10:
        return f"200{Year:.0f}"
    elif Year <= 23:
        return f"20{Year:.0f}"
    elif Year < 100:
        return f"19{Year:.0f}"
    else:
        return f"{Year:.0f}"

df.Year = df.Year.apply(lambda x: fix_Year(x))
#f스트링을 사용하여 10보다 작은숫자는 200을, 23보다 작거나 같으면 20을, 100보다 작은 수에 대해서는 19를 앞에 붙여주었다
````

````python
# 5년 단위 연도별 Global_sales의 막대그래프
df_5year_sales = df.groupby(df['Year']//5*5)['Global_sales'].sum() #//5*5를 사용해 5년단위로 묶어서 더할 수 있다
# 2015년 이후 데이터는 너무 작아서 2010에 합쳐주었다
df_5year_sales_new = df.groupby(df['Year'].apply(lambda x: 2010 if x >= 2015 else x//5*5))['Global_sales'].sum()

# 막대 그래프를 그립니다.
plt.figure(figsize=(10,6))
colors = plt.cm.Pastel1(range(len(df_5year_sales_new)))
sns.barplot(x=df_5year_sales_new.index, y=df_5year_sales_new.values, palette=colors)
plt.title('Global_sales', fontweight='bold', fontsize=24)
plt.xlabel('Year')
plt.ylabel('Sales (millions)')
plt.show()
````

````python
#세대별 인기게임
bins = [1980, 1990, 2000, 2010, 2020]
labels = ['1980s', '1990s', '2000s', '2010s']
df['Decade'] = pd.cut(df['Year'], bins=bins, labels=labels)
df_top10_decade = df.groupby(['Decade']).apply(lambda x: x.nlargest(100,'Global_sales')).reset_index(drop=True)
df_top10_decade[['Decade', 'Name', 'Global_sales']]
#유사게임이 상당히 많다
````

Miscellaneous(잡다한) 게임 장르는 일반적으로 특정한 장르에 귀속되지 않는 게임을 의미합니다. 예를 들면, 퀴즈 게임, 퍼즐 게임, 탈출 게임 등이 있습니다. 이는 다른 게임 장르와 구분하기 어렵거나 혼합된 장르를 포함하기도 합니다.

Platform(플랫폼) 게임 장르는 일반적으로 플랫폼을 오르내리며, 여러 장애물을 피하거나 적을 물리치며 스테이지를 클리어하는 게임을 의미합니다. 예를 들어, 슈퍼 마리오 브라더스, 소닉 더 헤지혹, 레이맨 등이 있습니다. 이는 다른 장르와 결합하여, 플랫폼 요소를 추가한 메타 게임도 존재합니다.

````python
# 같은 게임끼리 그룹화하고, 그룹에 속한 게임 수를 세어 game_count 칼럼에 저장
df_gc = df.groupby('Name').agg({'Genre':'first'})

# game_count 칼럼을 기준으로 내림차순 정렬
df_gc = df_gc.assign(game_count=df.groupby('Name').size()).sort_values(by='game_count', ascending=False)

# 결과 출력
df_gc.head(15)

#작품 하나가 메가히트 달성시, 다수의 플랫폼에서 여러 버전으로 출시하는경우가 많다
#이러한 메가히트작의 경우, 후속작, 모방게임이 트렌드와 수익성을 고려하지않고 마구잡이로 생성이 되는경우가 많고
#가장 보편적이고 대중적인 액션, 스포츠 장르에 많이 분포되어 있다
````
